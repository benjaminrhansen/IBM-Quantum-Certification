```python
%matplotlib inline
%config InlineBackend.print_figure_kwargs={'facecolor' : "w"}
from qiskit import *
from qiskit.visualization import *
from qiskit import QuantumCircuit
from qiskit.providers.aer import QasmSimulator
from qiskit.quantum_info import Statevector
import numpy
from math import cos, sin, pi
from IPython import display
```

```python
display.Image("../Sample Question 2.png")
```

```python
qc = QuantumCircuit(1) # qubit parameter
qc.ry(3 * pi/4, 0) # q0
```

```python
# theoretical final quantum results
statevector = Statevector(qc)
plot_bloch_multivector(statevector)
```

```python
qc.measure_all()
qc.draw('mpl')
```

```python
sim = QasmSimulator(method="statevector")
job = execute(qc, backend=sim, shots=1024)
result = job.result()
counts = result.get_counts(qc)
counts
```

```python
print("Probability of measuring 0:",counts['0']/1024)
plot_histogram(counts)
```
